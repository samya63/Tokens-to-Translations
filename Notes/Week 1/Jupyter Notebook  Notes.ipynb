### <span style="color:green"> Important </span>
1. The order of execution of cells may not be from top to bottom
2. The number in square brackets on the left of each code box denotes the order of running of the code boxes
3. It is important to execute the codes in such an order so that the codes execute properly
4. The extension for jupyter files is .ipynb
5. Jupyter notebooks can be exported in other formats like pdf and html
6. Jupyter notebooks can also be created on cloud environments like Google CoLab and Kaggle
7. In Kaggle we get access to all the datasets they have

### <span style="color:green">Some Keyboard Shortcuts</span>
1. **Shift + Enter** - Runs a cell and goes to next cell
2. **Ctrl + Enter** - Runs a cell and stays in that cell
3. **Arrow Keys** - Move cells
4. **ESC** - To escape out of a cell
5. **A** - Add a cell above the cell you escaped out of.
6. **B** - Add a cell below the cell you escaped out of.
7. **J and K** - Move cells.
