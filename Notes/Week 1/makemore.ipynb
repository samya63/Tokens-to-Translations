### <span style="color:green"> What is makemore? </span>
It makes more of what data it receives. It is a character level language model. It is modelling a sequence of characters and learns to predict the next character in the sequence.

```python
words=open('names.txt','r').read().splitlines()
```

#### <span style="color:green"> What information do we get from each word?</span>
eg.olivia
1. 'l' is likely to follow 'o'; 'i' is likely to follow 'ol'; and so on
2. The word is likely to end at 'olivia'

#### <span style="color:green"> What is a bigram language model?</span>
A model that predicts the next token only using the previous token.
Very simple and weak language model.


```python
import torch

#initialising the data array as well as the mappings
letters=sorted(list(set("".join(words))))
n=len(letters)
P=torch.zeros((n+1,n+1),dtype=int)
stoi={ch:i+1 for i,ch in enumerate(letters)}
stoi['.']=0
itos={i:ch for ch,i in stoi.items()}
for word in words:
    chs="."+word+"."
    for ch1,ch2 in zip(chs,chs[1:]):
        P[stoi[ch1],stoi[ch2]]+=1

#normalising the data arrray
P=P/torch.sum(P,1,keepdim=True)
```

Be careful that torch deals with ***Tensor*** values, so we ust get the value inside them using ***.items()*** 

#### <span style="color:green"> torch.sum(tensor,dim,keepdim) </span>
1. dim: is the dimension along which the summing occurs
2. keemdims=True if you want to keep that dimension as a 1 in the shape attribute, otherwise that dimension just disappears

#### <span style="color:green"> What is Broadcasting? </span>
Even if shapes dont match two tensors can be operate dwith each other by copying data in the non matching dimensions


#### <span style="color:green"> Broadcasting Rules </span>
1. Compare shape from *****right***** to *****left*****
2. For each dimension, the sizes are compatible if they are equal or they are 1 or one tensor has no dimension there treated like leading 1s.

#### <span style="color:green"> torch.multinomial(input, num_samples,replacement=False,*,generator=None,out=None)</span>
Returns a tensor where each row contains num_sample indices sampled fro the multinomial probability distribution located in the corresponding row of tensor input
1. **input** is the probability distribution
2. If **replacement** is **True** then the samples are drawn with replacement.
3. When drawn without replacement **num_samples** must be lower than or equal to number of non-zero elements in **input**(or the minimum number of non-zero elements in each row of **input** if it is a matrix)
4. If **input** is a vector, **out** is a vector of size **num_samples**
5. If **input** is a matrix with m rows then, **out** is a matrix of shape **m** x **num_samples**





#### <span style="color:green">class torch.Generator(device='cpu')</span>
It is an object controlling the random object generator state in PyTorch <br>
g=torch.Generator()<br>
g.manual_seed(*some_number*)


```python
g=torch.Generator().manual_seed(2147483647)
torch.multinomial(P[0],1,replacement=True,generator=g)
for i in range(10):
    word=itos[torch.multinomial(P[0],1,replacement=True,generator=g).item()]
    while word[-1]!='.':
        word+=itos[torch.multinomial(P[stoi[word[-1]]],1,replacement=True,generator=g).item()]
    print(word)
```
